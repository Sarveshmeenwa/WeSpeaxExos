```python
import pandas as pd
import spacy
from lemminflect import getAllInflections
import random

nlp = spacy.load("en_core_web_sm")
```

```python
japonaisCorpus = pd.read_csv("../Datasets/JaponaisCorpus.tsv", sep="\t", encoding="utf-8").dropna()
japonaisCorpus
```

```python
for sentence in japonaisCorpus["minSource"][:10]:
    doc = nlp(sentence)
    for i, token in enumerate(doc):
        if token.pos_ == "VERB":
            inflections = list(getAllInflections(token.lemma_, upos="VERB").values())
            s = sentence.replace(token.text, random.choice(inflections)[0])
    print([sentence, s])
```
